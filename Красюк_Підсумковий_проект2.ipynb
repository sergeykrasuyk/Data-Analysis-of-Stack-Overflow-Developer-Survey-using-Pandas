**Завдання 7. Географічний аналіз компенсації Python-розробників**



```python
# Обираємо потрібні колонки
df_python = df_public[has_python][['Country', 'ConvertedCompYearly']]
```

```python
# Очищаємо від пропусків
df_python = df_python.dropna(subset=['ConvertedCompYearly', 'Country'])
```

```python
# Виконуємо групування та агрегацію(знаходимо середнє та медіану)
result = (
    df_python
    .groupby('Country')['ConvertedCompYearly']
    .agg(['mean', 'median'])
    .reset_index()
)
```

```python
result.columns = ['Country', 'avg_compensation', 'median_compensation']
```

```python
# Сортуємо за спаданням
result = result.sort_values(by='avg_compensation', ascending=False)
print(result)
```

**Завдання 9. Аналіз популярності Python по віковим категоріям**

```python
df_public['Age'].describe()
```

```python
df_public['AgeGroup'] = df_public['Age']

python_by_age = (
    df_public
    .assign(has_python=has_python)
    .groupby('AgeGroup')['has_python']
    .mean()
    .mul(100)
    .reset_index(name='python_percent')
)

python_by_age = python_by_age.sort_values(by='python_percent', ascending=False)
print(python_by_age.round({'python_percent': 1}))
```
